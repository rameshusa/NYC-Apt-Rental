   ### Box Plot

```python
# limit number of bedrooms and prices
filtered_train =  dftrain[(dftrain.price<8000) & (dftrain.bedrooms<=4)]
palette = {"high": "#167ec9", "medium":"#82d147", "low":"#a1a59f"}
```

```python
plt.figure(figsize=(20,10))
sns.boxplot(x="bedrooms", y="price", hue="interest_level", data=filtered_train, palette=palette, )
```

### Conclusion: Interest level (All - Low, Medium, High) increases as the number of Bedroom increases
### Rent of $4000+ Requires more than 3 Bedrooms

```python
plt.figure(figsize=(20,10))
sns.boxplot(x="bathrooms", y="price", hue="interest_level", data=filtered_train, palette=palette)
```


### Conclusion: Interest level (All - Low, Medium, High) increases as the number of Bathroom increases, However Bathroom size of 1 & 2 are prefered most. Rent of $5000 requires more than 3 bathrooms.



### Scatter plot

```python
plt.figure(figsize=(20,10))
for interest in ['low', 'medium', 'high']:
    plt.scatter(filtered_train[filtered_train.interest_level==interest].bedrooms, 
                filtered_train[filtered_train.interest_level==interest].price, 
                c=palette[interest])
```

#### Lot of interest is for 0-2 bedrooms with less than $4000 Rent. 

```python
plt.figure(figsize=(20,10))
for interest in ['low', 'medium', 'high']:
    plt.scatter(filtered_train[filtered_train.interest_level==interest].bathrooms, 
                filtered_train[filtered_train.interest_level==interest].price, 
                c=palette[interest])
```

#### Lot of interest is for 1 - 2 Bathroom listings
